```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.describe()
```

```python
tips['time'].value_counts()
```

```python
sns.barplot(x='time',y='total_bill',data=tips,hue='smoker')  #visualization of a group by action
```

```python
tips['sex'].value_counts()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator =np.std, hue='smoker')  #visualization of a group by action
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)  #visualization of a group by action
```

```python
sns.countplot(x='time' ,data=tips) #number of each value
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker') #people that smaoke in the days 
```

```python
sns.violinplot(x='sex',y='total_bill', data =tips, hue='smoker', split=True)
```

```python
sns.stripplot(x='day',y='total_bill', data =tips, hue='sex',split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
#combination of many plots
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)  #combining two plots
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')  #used for EDA
```

```python
#factor plot
sns.factorplot(x='day',y='total_bill',data=tips, kind='bar')  #can be used to recreate other plots
```
